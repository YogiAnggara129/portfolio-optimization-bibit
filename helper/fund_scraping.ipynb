## Prepare Web Driver

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

driver = webdriver.Chrome('../driver/chromedriver.exe')

```

## Login to pasardana.id

```python
with open('../data/user_pasardana.json') as f:
    user_pasardana = json.load(f)

driver.get('https://www.pasardana.id/login')
driver.find_element(
    by=By.XPATH, value='//*[@id="EmailAddress"]').send_keys(user_pasardana['email'])
driver.find_element(
    by=By.XPATH, value='//*[@id="Password"]').send_keys(user_pasardana['password'])
driver.find_element(
    by=By.XPATH, value='/html/body/div/main/div/div/div/section/form/div[4]/div/button').click()
```

## Scraping NAV History

```python
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support import expected_conditions as EC
```

```python
df_bibit_fund_desc = pd.read_csv('../data/fund_desc_in_bibit.csv')
bibit_fund_name = df_bibit_fund_desc['name']

for name in bibit_fund_name[86:]:
    delay_search = 3
    delay_history = 6

    # go to search fund path
    print(f'Scraping: {name}')
    driver.get('https://www.pasardana.id/fund/search')
    time.sleep(delay_search)
    driver.find_element(
        by=By.XPATH, value='/html/body/div/ui-view/ui-view/ui-view/main/div[1]/div/div[2]/div[3]/div[2]/form/div/input').send_keys(name)
    time.sleep(delay_search)
    driver.find_element(
        by=By.XPATH, value='//*[@id="jqxWidget"]/div[4]/div[2]/table/tbody/tr[1]/td[3]/a').click()

    try:
        WebDriverWait(driver, delay_history).until(EC.presence_of_element_located(
            (By.XPATH, '//*[text()="Download CSV"]'))).click()
        print("- Success!")
    except TimeoutException:
        print("- Loading took too much time!")
```

```python
driver.quit()
```

# Data Cleansing

```python
import os
import pandas as pd
```

## List NAV file name

```python
nav_dir = '../data/nav/'
nav_file_list = os.listdir(nav_dir)
```

## Cleansing NAV Datasets

```python
def cleansing_nav(df):
    for i in range(len(df)):
        if pd.isna(df.iloc[i,1]):
            df.iloc[i,1] = df.iloc[i,2]
    df.iloc[:,0] = pd.to_datetime(df.iloc[:,0])
    df = df.iloc[:,[0,1]]
    df = df[df['DateTime'] >= pd.to_datetime('2021-01-01')]
    df = df.set_index('DateTime').resample('D', origin='2021-01-01').ffill()
    df = df.fillna(method='ffill')
    # df = df.iloc[list(range(0,len(df.index),7)),:]
    return df
```

```python
is_first = True
for file in nav_file_list:
    if is_first:
        df_nav_bibit = pd.read_csv(nav_dir + file)
        df_nav_bibit = cleansing_nav(df_nav_bibit)
        is_first = False
        continue
    df_temp = pd.read_csv(nav_dir + file)
    df_temp = cleansing_nav(df_temp)
    # df_nav_bibit = pd.merge(df_nav_bibit, df_temp, left_index=True, right_index=True)
    # df_nav_bibit.join(df_temp)
    df_nav_bibit = pd.concat([df_nav_bibit, df_temp], axis=1)

df_nav_bibit.head()
```

```python
df_nav_bibit = df_nav_bibit.iloc[list(range(0,len(df_nav_bibit.index),7)), :]
df_nav_bibit = df_nav_bibit.iloc[1:,:]
df_nav_bibit.head()
```
